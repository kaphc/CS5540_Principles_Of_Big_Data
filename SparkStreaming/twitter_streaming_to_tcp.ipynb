```python
import socket
import sys
import requests
import requests_oauthlib
import json
import Credentials.twitter_credentials
```

```python
from google.cloud import language_v1
from google.cloud.language_v1 import enums
```

```python
CONSUMER_KEY = Credentials.twitter_credentials.CONSUMER_KEY
CONSUMER_SECRET = Credentials.twitter_credentials.CONSUMER_SECRET
ACCESS_TOKEN = Credentials.twitter_credentials.ACCESS_TOKEN
ACCESS_SECRET = Credentials.twitter_credentials.ACCESS_SECRET
my_auth = requests_oauthlib.OAuth1(CONSUMER_KEY, CONSUMER_SECRET,ACCESS_TOKEN, ACCESS_SECRET)
```

```python
client = language_v1.LanguageServiceClient()
type_ = enums.Document.Type.PLAIN_TEXT
```

```python
def get_track():
    f = open('data/stopwords.txt')
    query = ""
    for line in f:
        line = line.strip("\n")
        query = query + "," + line
    return query[4:]    
```

```python
def get_tweets():
    url = 'https://stream.twitter.com/1.1/statuses/filter.json'
    query_data = [('language', 'en'), ('track', get_track())]
    query_url = url + '?' + '&'.join([str(t[0]) + '=' + str(t[1]) for t in query_data])
    response = requests.get(query_url, auth=my_auth, stream=True)
    print(query_url, response)
    return response
```

```python
def send_tweets_to_spark(http_resp, tcp_connection):
    for line in http_resp.iter_lines():
        try:
            full_tweet = json.loads(line)
            tweet_text = full_tweet['text']
            if(len(tweet_text.split()) > 20):
                if(classify_tweet(tweet_text)):
                    print("Tweet Text: " + tweet_text)
                    print ("------------------------------------------")
                    tweet_data = bytes('{' + 'tweet_text' + tweet_text + '}' + "/n", "utf-8")
                    tcp_connection.send(tweet_data)
        except:
            e = sys.exc_info()[0]
            print("Error: %s" % e)
```

```python
def analyze_tweet(text_content):
    type_ = enums.Document.Type.PLAIN_TEXT
    language = "en"
    document = {"content": text_content, "type": type_, "language": language}
    encoding_type = enums.EncodingType.UTF8

    response = client.analyze_sentiment(document, encoding_type=encoding_type)
    return str(response.document_sentiment.score)
```

```python
def classify_tweet(text_content):
    language = "en"
    document = {"content": text_content, "type": type_, "language": language}

    response = client.classify_text(document)
    for category in response.categories:
        if(category.name.startswith("/Health") or category.name.startswith("/News/Health News")):
            return True
    return False
```

```python
TCP_IP = "localhost"
TCP_PORT = 9010
conn = None
```

```python
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.bind((TCP_IP, TCP_PORT))
s.listen(1)
print("Waiting for TCP connection...")
conn, addr = s.accept()
print("Connected... Starting getting tweets.")
resp = get_tweets()
send_tweets_to_spark(resp, conn)
```
